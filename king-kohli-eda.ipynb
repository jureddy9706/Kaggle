```python
import pandas as pd
import plotly.express as px
```

```python
df = pd.read_csv('../input/virat-kohlis-71-centuries/71 Centuries of Virat Kohli.csv')
df.head()
```

```python
df['Unnamed: 14'].isnull().sum()
```

let's remove unnamed :14, Strike rate and date because all are null values and rename Column1 as Match Place 

```python
df.drop('Unnamed: 14',inplace=True,axis=1)
df.rename(columns={'Column1':'Match Place'},inplace=True)
df.drop('Strike Rate',inplace=True,axis=1)
df.drop('Date',inplace=True,axis=1)
```

```python
df.info() # FINALLY NO NULL VALUES
```

# UNIVARIATE ANALYSIS

# 1.Highest Score

```python
df.sort_values(by='Score',ascending=False).head(1)
```

**HIGHEST SCORE BY KOHLI IS 254 AND IT'S AGAINST SOUTH AFRICA IN A TEST MATCH AS A CAPTAIN AND INDIA WON THE MATCH**

# 1.1 Most Centuries

```python
df['Venue'].value_counts().head(5)
```

```python
fig=px.bar(x=df['Venue'].value_counts().head(5).index,
       y=df['Venue'].value_counts().head(5).values,
       text=df['Venue'].value_counts().head(5).values,
       color=df['Venue'].value_counts().head(5).index,
       title='TOP 5 MOST CENTURIES IN STADIUM',
       template='plotly_dark')
fig.update_layout(xaxis_title='STADIUM',yaxis_title='NO.OF CENTURIES')
```

**OUT OF 71 CENTURIES MAXIMUM CENTURIES ARE IN ADELAIDE OVAL STADIUM**

# 1.2 Most Centuries against

```python
fig=px.bar(x=df['Against'].value_counts().index,
       y=df['Against'].value_counts().values,
       text=df['Against'].value_counts().values,
       color=df['Against'].value_counts().index,
       title='MOST CENTURIES AGAINST',
       template='plotly_dark')
fig.update_layout(xaxis_title='COUNTRY',yaxis_title='NO.OF CENTURIES')
```

**HE HAD SCORE MORE CENTURIES AGAINST AUSTRALIA**

```python
df['Out/Not Out'].value_counts()
```

# 1.3 OUT VS NOT OUT

```python
fig=px.pie(names=df['Out/Not Out'].value_counts().index,
           values=df['Out/Not Out'].value_counts().values,
           title='OUT VS NOT OUT',
            template='plotly_dark')
fig.update_traces(textinfo='label+percent+value')
fig.show()
```

```python
fig=px.pie(names=df['Batting Order'].value_counts().index,
           values=df['Batting Order'].value_counts().values,
           title='Batting Order',
            template='plotly_dark')
fig.update_traces(textinfo='label+percent+value')
fig.show()
```

**HALF THE CENTURIES ARE SCORED IN PLAYING AT 3RD ORDER**

```python
fig=px.pie(names=df['Man of the Match'].value_counts().index,
           values=df['Man of the Match'].value_counts().values,
           title='Man of the Match',
            template='plotly_dark')
fig.update_traces(textinfo='label+percent+value')
fig.show()
```

**MORE THAN HALF TIMES KOHLI WON THE MAN OF THE MATCH**
