# 1.‚òûIMPORTING LIBRARIES

```python
import pandas as pd
import numpy as np
import plotly.express as px
```

```python
df=pd.read_csv('../input/data-science-job-salaries/ds_salaries.csv')
df.head()
```

**WE DON'T NEED 'UNNAMED:0','SALARY','SALARY_CURRENCY' BECAUSE WE ARE GONNA TAKE SALARY I USD SO LET'S REMOVE THEM**

```python
df.drop(['Unnamed: 0','salary','salary_currency'],axis=1,inplace=True)
df.head()
```

```python
df.info()
```

**NO NULL VALUES**

```python
df.duplicated().value_counts()
```

**THERE ARE 42 DUPLICATED VALUES SO LET'S REMOVE THEM**

```python
df.drop_duplicates(inplace=True)
```

```python
df.duplicated().value_counts() # SO, all duplicates are removed
```

**CONVERT THE 2-LETTER COUNTRY NAME TO FULL NAMES FOR OUR CONVENIENCE**

```python
pip install country_converter --upgrade
```

```python
import country_converter as coco #THIS WILL CONVERT OUR NAMES 
```

**Converting company_location and employee_residence**

```python
df['company_location']=coco.convert(names=df['company_location'],to='name_short')
df['employee_residence']=coco.convert(names=df['employee_residence'],to='name_short')
 
```

**REPLACING THE TEXT OUR CONVENIENCE**

```python
df['experience_level']=df['experience_level'].replace(['MI','SE','EX','EN'],[' Mid-level','Senior-level','Executive-level','Entry-level'])
df['employment_type']=df['employment_type'].replace(['FT','CT','PT','FL'],['Full-time','Contract','Part-time','Freelance'])
df['company_size']=df['company_size'].replace(['S','L','M',],['Small','Large','Medium'])
df['remote_ratio']=df['remote_ratio'].replace([0,50,100],['Onsite','Hybrid','Remote'])
```

```python
df.head()
```

# 2.JOBS ANALYSIS

```python
experience=df['experience_level'].value_counts()
fig=px.pie(names=experience.index,
           values=experience.values,
           color=experience.index,
           title='Jobs based on Experience level',
           template='plotly_dark')
fig.update_traces(textinfo='label+percent+value')
fig.show()
```

```python
remote=df['remote_ratio'].value_counts()
fig=px.pie(names=remote.index,
           values=remote.values,
           color=remote.index,
           title='Jobs based on remote',
           template='plotly_dark')
fig.update_traces(textinfo='label+percent+value')
fig.show()
```

**AS WE CAN SEE MANY COMPANIES ARE DOING WORK FROM HOME WHICH IS 61.2%**

# COMAPNIES ANLYSIS

```python
size=df['company_size'].value_counts()
fig=px.pie(names=size.index,
           values=size.values,
           color=size.index,
           title='COMPANIES BASED ON SIZE',
           template='plotly_dark')
fig.update_traces(textinfo='label+percent+value')
fig.show()
```

**MEDIUM SIZED COMPANIES ARE MORE THAN HALF IN THE DATA SCIENCE INDUSTRY**

```python
df['same area']=df.apply(lambda x: 'same location' if x.employee_residence==x.company_location else 'different location',axis=1)
df['same area'].value_counts()
```

**OUT OF 555 EMPLOYEES 514 ARE RESIDING IN THE SAME STATE AS THE COMPANY LOCATION**

# EMPLOYEES IN SAME AREA BUT WORKING FULLY REMOTE

```python
df['same area but remote']=df.apply(lambda x: 'yes' if (x.employee_residence==x.company_location)&(x.remote_ratio=='Remote') else 'no',axis=1)
remotly=df['same area but remote'].value_counts()
fig=px.pie(names=remotly.index,
           values=remotly.values,
           color=remotly.index,
           title='EMPLOYEES IN SAME AREA BUT WORKING FULLY REMOTE',
           template='plotly_dark')
fig.update_traces(textinfo='label+percent+value')
fig.show()
```

**EVEN THOUGH 308 LIVING IN THE SAME AREA AS IN THE COMPANY'S LOCATION BUT THEY ARE WORKING FULLY REMOTE I THINK COVID PLAYS MAJOR ROLE IN THIS.**
